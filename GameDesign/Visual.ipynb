```python
import numpy as np
import matplotlib.pyplot as plt


base_req = 100
growth_rate_exp = 1.1
cons = 2
levels = np.arange(1,100)

req_exp = base_req * growth_rate_exp ** (levels - 1)

req_poly = base_req * levels ** cons

plt.figure(figsize=(10,5))
plt.plot(levels, req_exp, marker='o', label='Exponential Growth')
plt.plot(levels, req_poly, marker='x', label='Polynomial Growth')
plt.title("Levbel Progression")
plt.xlabel("Level")
plt.ylabel("Experience Required")
plt.legend()
plt.grid()
plt.show()
```
